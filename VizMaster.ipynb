# corr() Notes

*standard correlation coefficient (Pearson's r)* : A numerical summary of the strength of the linear association between the variables. 

source : 
    1. http://www.people.vcu.edu/~pdattalo/706SuppRead/Pearson%27s%20r.html

```python
#build correlation matrix with corr()
corr_matrix = df.corr()

#view a specific feature's correlation with other features
feature_name = "Survived"
corr_matrix[feature_name].sort_values(ascending=False)
```

# Visualizing Correlation Matrix using shaded cells

*Filter column names to only numeric datatypess. Corr() only evaluates numeric datatypes

```python
import matplotlib
import matplotlib.pyplot as plt

plt.matshow(corr_matrix)
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.colorbar()
plt.show()
```
